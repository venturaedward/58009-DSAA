```python
import matplotlib.pyplot as plt

def plot_polygon(coordinates, name, num_sides, area):
    coordinates.append(coordinates[0])  # Repeat the first point to create a 'closed loop'
    xs, ys = zip(*coordinates)  # Create lists of x and y values

    plt.figure()
    plt.plot(xs, ys)
    plt.title(f"Polygon: {name}\nNumber of Sides: {num_sides}\nArea: {area}")
    plt.show()

def get_polygon_input(shape):
    if shape == "square":
        num_sides = 4
        side_length = float(input("Enter the side length of the square: "))
        area = side_length ** 2
        coordinates = [[0, 0], [side_length, 0], [side_length, side_length], [0, side_length]]
    elif shape == "rectangle":
        num_sides = 4
        width = float(input("Enter the width of the rectangle: "))
        height = float(input("Enter the height of the rectangle: "))
        area = width * height
        coordinates = [[0, 0], [width, 0], [width, height], [0, height]]
    elif shape == "triangle":
        num_sides = 3
        base = float(input("Enter the base length of the triangle: "))
        height = float(input("Enter the height of the triangle: "))
        area = 0.5 * base * height
        coordinates = [[0, 0], [base, 0], [base / 2, height]]
    else:
        print("Invalid shape!")
        return None, 0, 0
    return coordinates, num_sides, area

# Example usage
shape = input("Enter the shape of the polygon (square, rectangle, triangle): ").lower()
polygon_coordinates, polygon_num_sides, polygon_area = get_polygon_input(shape)
polygon_name = f"{shape.capitalize()}"

if polygon_coordinates:
    plot_polygon(polygon_coordinates, polygon_name, polygon_num_sides, polygon_area)
```
